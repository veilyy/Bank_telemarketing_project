```python
import pandas as pd
import numpy as np
import joblib
import json
```

```python
df = pd.read_csv('../data/raw/bank-additional-full.csv', sep=';')
```

```python
y = (df['y'] == 'yes').astype(int)
```

```python
X = df.drop(columns = 'y')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 42)
```

```python
# Импортируем функции обучения модели
from src.preprocessing import *
```

```python
Preproc = Preprocessor(use_scaler = False, duration = True)
```

```python
X_train_p = Preproc.fit_transform(X_train)

X_test_p= Preproc.transform(X_test)
```

```python
X_train_p
```

```python
import optuna
import xgboost as xgb
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score, classification_report
import matplotlib.pyplot as plt

# Optuna objective
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'gamma': trial.suggest_float('gamma', 0, 2),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-8, 1.0, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-8, 1.0, log=True),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 2, 8),
        'objective': 'binary:logistic',
        'random_state': 42,
        'n_jobs': -1,
        'verbosity': 0
    }
    
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    f1_scores = []
    
    for train_idx, val_idx in cv.split(X_train_p, y_train):
        X_fold_train = X_train_p.iloc[train_idx]
        y_fold_train = y_train.iloc[train_idx]
        X_fold_val = X_train_p.iloc[val_idx]
        y_fold_val = y_train.iloc[val_idx]
        
        model = xgb.XGBClassifier(**params)
        model.fit(X_fold_train, y_fold_train, verbose=False)
        
        y_pred = model.predict(X_fold_val)
        f1 = f1_score(y_fold_val, y_pred, zero_division=0)
        f1_scores.append(f1)
    
    return np.mean(f1_scores)

# Optuna optimization
study = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler(seed=42))
study.optimize(objective, n_trials=150, show_progress_bar=True)

print(f"Best F1 (CV): {study.best_value:.4f}")

# Train final model
best_params = study.best_params.copy()
best_params.update({
    'objective': 'binary:logistic',
    'random_state': 42,
    'n_jobs': -1,
    'verbosity': 0
})

final_model = xgb.XGBClassifier(**best_params)
final_model.fit(X_train_p, y_train, verbose=False)

# Predictions
y_pred = final_model.predict(X_test_p)
y_pred_proba = final_model.predict_proba(X_test_p)[:, 1]

# Metrics
print("\n" + "="*40)
print("📊 TEST METRICS")
print("="*40)
print(classification_report(y_test, y_pred, digits=4))

# Feature importance
feature_importance = pd.DataFrame({
    'feature': X_train_p.columns,
    'importance': final_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\n🎯 TOP-10 FEATURES:")
print(feature_importance.head(10).to_string(index=False))

# 📊 ANALYTICS & VISUALIZATION
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Feature importance
top_n = 15
axes[0, 0].barh(range(top_n), feature_importance['importance'].head(top_n))
axes[0, 0].set_yticks(range(top_n))
axes[0, 0].set_yticklabels(feature_importance['feature'].head(top_n))
axes[0, 0].set_xlabel('Importance')
axes[0, 0].set_title('Top Feature Importance')
axes[0, 0].invert_yaxis()

# 2. Optuna optimization history
axes[0, 1].plot([t.number for t in study.trials], 
                [t.value for t in study.trials], 
                marker='o', alpha=0.5)
axes[0, 1].axhline(y=study.best_value, color='red', linestyle='--')
axes[0, 1].set_xlabel('Trial')
axes[0, 1].set_ylabel('F1 Score')
axes[0, 1].set_title(f'Optuna History (Best: {study.best_value:.4f})')
axes[0, 1].grid(True, alpha=0.3)

# 3. Parameter importance
optuna.visualization.matplotlib.plot_param_importances(study, ax=axes[1, 0])
axes[1, 0].set_title('Parameter Importance')

# 4. Probability distribution
axes[1, 1].hist(y_pred_proba[y_test == 0], bins=30, alpha=0.5, label='Class 0', density=True)
axes[1, 1].hist(y_pred_proba[y_test == 1], bins=30, alpha=0.5, label='Class 1', density=True)
axes[1, 1].axvline(x=0.5, color='red', linestyle='--', label='Threshold 0.5')
axes[1, 1].set_xlabel('Predicted Probability')
axes[1, 1].set_ylabel('Density')
axes[1, 1].set_title('Probability Distribution')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 📈 SUMMARY
print("\n" + "="*40)
print("📋 SUMMARY")
print("="*40)
print(f"• CV F1: {study.best_value:.4f}")
print(f"• Test F1: {f1_score(y_test, y_pred):.4f}")
print(f"• Test AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")
print(f"• Top feature: {feature_importance.iloc[0]['feature']} ({feature_importance.iloc[0]['importance']:.3f})")
print(f"• Key params: n_estimators={best_params.get('n_estimators')}, max_depth={best_params.get('max_depth')}")
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report, f1_score, roc_auc_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

# Ваши лучшие параметры
best_params = {
    'colsample_bytree': 0.7146147760276441,
    'gamma': 0.9738824124939458,
    'learning_rate': 0.01401492606384062,
    'max_depth': 6,
    'min_child_weight': 5,
    'n_estimators': 118,
    'n_jobs': -1,
    'objective': 'binary:logistic',
    'random_state': 42,
    'reg_alpha': 4.764407281074392e-08,
    'reg_lambda': 0.0008900337922885305,
    'scale_pos_weight': 5.168721411436508,
    'subsample': 0.7406799845907971,
    'verbosity': 0,
}

print("="*50)
print("🎯 ОБУЧЕНИЕ XGBOOST С ОПТИМАЛЬНЫМИ ПАРАМЕТРАМИ")
print("="*50)
print(f"Используем {X_train_p.shape[1]} фич")
print(f"Размер train: {X_train_p.shape}")
print(f"Размер test: {X_test_p.shape}")

# 1. Обучение модели
print("\n⚙️ Обучение модели...")
model = xgb.XGBClassifier(**best_params)
model.fit(X_train_p, y_train, verbose=False)
print("✅ Модель обучена")

# 2. Предсказания
y_pred = model.predict(X_test_p)
y_pred_proba = model.predict_proba(X_test_p)[:, 1]

# 3. Метрики
print("\n" + "="*50)
print("📊 МЕТРИКИ НА ТЕСТЕ (порог 0.5)")
print("="*50)
print(classification_report(y_test, y_pred, digits=4))

print(f"\n📈 Дополнительные метрики:")
print(f"F1 Score: {f1_score(y_test, y_pred):.4f}")
print(f"AUC-ROC: {roc_auc_score(y_test, y_pred_proba):.4f}")

# 4. Матрица ошибок
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()
print(f"\n📊 Матрица ошибок:")
print(f"TN={tn}, FP={fp}")
print(f"FN={fn}, TP={tp}")
print(f"Accuracy: {(tp+tn)/(tp+tn+fp+fn):.4f}")
print(f"% предсказанных 1: {y_pred.mean():.1%}")

# 5. Важность фич
feature_importance = pd.DataFrame({
    'feature': X_train_p.columns,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

print("\n" + "="*50)
print("🎯 ТОП-15 ВАЖНЕЙШИХ ФИЧ")
print("="*50)
print(feature_importance.head(15).to_string(index=False))

# 6. Быстрая проверка разных порогов
print("\n" + "="*50)
print("🔧 БЫСТРАЯ ПРОВЕРКА ПОРОГОВ")
print("="*50)

thresholds = [0.3, 0.4, 0.5, 0.6, 0.7]
print("\nПорог | F1     | Recall | Precision | %1")
print("-" * 45)

for thresh in thresholds:
    y_pred_thresh = (y_pred_proba >= thresh).astype(int)
    f1 = f1_score(y_test, y_pred_thresh)
    recall = recall_score(y_test, y_pred_thresh)
    precision = precision_score(y_test, y_pred_thresh)
    pct_1 = y_pred_thresh.mean()
    print(f"{thresh:.1f}   | {f1:.4f} | {recall:.4f} | {precision:.4f}   | {pct_1:.1%}")

# 7. Визуализация
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Важность фич
top_n = 15
axes[0].barh(range(top_n), feature_importance['importance'].head(top_n))
axes[0].set_yticks(range(top_n))
axes[0].set_yticklabels(feature_importance['feature'].head(top_n))
axes[0].set_xlabel('Importance')
axes[0].set_title('Top Feature Importance')
axes[0].invert_yaxis()

# Распределение вероятностей
axes[1].hist(y_pred_proba[y_test == 0], bins=30, alpha=0.5, label='Class 0', density=True)
axes[1].hist(y_pred_proba[y_test == 1], bins=30, alpha=0.5, label='Class 1', density=True)
axes[1].axvline(x=0.5, color='red', linestyle='--', label='Threshold 0.5')
axes[1].set_xlabel('Predicted Probability')
axes[1].set_ylabel('Density')
axes[1].set_title('Probability Distribution')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# F1 по порогам
threshold_range = np.linspace(0.1, 0.9, 50)
f1_scores = []
for thresh in threshold_range:
    y_pred_temp = (y_pred_proba >= thresh).astype(int)
    f1_scores.append(f1_score(y_test, y_pred_temp))

axes[2].plot(threshold_range, f1_scores, linewidth=2)
axes[2].axvline(x=0.5, color='red', linestyle='--', label='Default=0.5')
axes[2].set_xlabel('Threshold')
axes[2].set_ylabel('F1 Score')
axes[2].set_title('F1 Score vs Threshold')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 8. Итог
print("\n" + "="*50)
print("📋 ИТОГОВЫЙ ОТЧЕТ")
print("="*50)
print(f"✅ Модель обучена с параметрами:")
print(f"   • n_estimators: {best_params['n_estimators']}")
print(f"   • max_depth: {best_params['max_depth']}")
print(f"   • learning_rate: {best_params['learning_rate']:.4f}")
print(f"   • scale_pos_weight: {best_params['scale_pos_weight']:.2f}")

print(f"\n📊 Результаты на тесте:")
print(f"   • F1 Score: {f1_score(y_test, y_pred):.4f}")
print(f"   • AUC-ROC: {roc_auc_score(y_test, y_pred_proba):.4f}")
print(f"   • Top feature: {feature_importance.iloc[0]['feature']}")
print(f"   • Важность top feature: {feature_importance.iloc[0]['importance']:.3f}")
```
